```python
import openai
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv())
client = openai.Client()
```

```python
resposta = client.chat.completions.create(
    model='gpt-4o',
    messages=[{
        'role': 'user',
        'content': [
            {'type': 'text', 'text': 'Descreva a imagem fornecida'},
            {'type': 'image_url', 'image_url': {'url': 'https://img.freepik.com/fotos-gratis/foto-de-uma-estrada-coberta-de-grama-e-varios-tipos-de-arvores_181624-30269.jpg'}}
        ]
    }]
)
print(resposta.choices[0].message.content)
```

```python
import base64

def encode_image(caminho_imagem):
    with open(caminho_imagem, 'rb') as img:
        return base64.b64encode(img.read()).decode('utf-8')
    
caminho = 'imagens/celulas.jpg'
base64_img = encode_image(caminho)

resposta = client.chat.completions.create(
    model='gpt-4o',
    messages=[{
        'role': 'user',
        'content': [
            {'type': 'text', 'text': 'Quantas celulas estão presentes na imagem?'},
            {'type': 'image_url', 'image_url': {
                'url': f'data:image/jpg;base64,{base64_img}'
            }}
        ]
    }],
    max_tokens=1000,
)
print(resposta.choices[0].message.content)
```
